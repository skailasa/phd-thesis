```python
%matplotlib inline

import matplotlib.pyplot as plt

import numpy as np
```

```python
def f1(n): return n**(4/3)
def f2(n): return n * np.log(n)
```

```python
d = 6

n = np.arange(8, 8**d)

plt.plot(n, f1(n), label='N^4/3')
plt.plot(n, f2(n), label='N log N')

plt.legend()
```

```python
plt.plot(n, f1(n)/f2(n), label='Mul Factor')
plt.legend()
```
