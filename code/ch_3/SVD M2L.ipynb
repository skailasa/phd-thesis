```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('svd_m2l.csv')
naive = df[df['method'] == 'n']
blocked = df[df['method'] == 'b']
```

```python
naive
```

```python
blocked
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

width = 0.75  # the width of the bars
multiplier = 0

offset = width * multiplier-0.35
rects = ax.bar(naive['order'] + offset, naive['runtime']/1e3, width, label='Naive', alpha=0.75)
# ax.bar_label(rects, padding=3)
multiplier += 1
offset = width * multiplier-0.35
rects = ax.bar(blocked['order'] + offset, blocked['runtime']/1e3, width, label='Blocked', alpha=0.75)

ax.set_xticks([5, 7, 9])
ax.set_xticklabels(['P=5', 'P=7', 'P=9'])
ax.legend(fontsize=16)

ax.grid(True, linestyle='--', which='major', color='grey', alpha=.75)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.set_xlabel('Expansion Order, $P$', size=20)
ax.set_ylabel('$T^{M2L}$ Runtime (s)', size=20)

fig.savefig(f'svd_m2l.png', dpi=300)
```
