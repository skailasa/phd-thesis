```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()

s = df[df['method']=='strumpack']
s_mv = df[df['method']=='strumpack_mv']
e = df[df['method']=='exafmm']
et = df[df['method']=='exafmmt']
f = df[df['method']=='fmm3d']
fl = df[df['method']=='flam']
fl_mv = df[df['method']=='flam_mv']
```

```python
fig, ax = plt.subplots(1,1)

ax.plot(f['N'], f['wall_time'], '--o', label='FMM3D')
ax.plot(fl['N'], fl['wall_time'], '--o', label='FLAM')
ax.plot(s['N'], s['wall_time'], '--o', label='STRUMPACK')
ax.plot(e['N'], e['wall_time'], '--x', label='ExaFMM')
ax.plot(et['N'], et['wall_time'], '--^', label='ExaFMM-T')


ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('N')
ax.set_ylabel('Time (s)')

ax.legend()
plt.savefig('../../images/ch_2/runtime.pdf', format='pdf')
```

```python
fig, ax = plt.subplots()

ax.plot(f['N'], f['wall_time'], '--o', label='FMM3D (for scale)')
ax.plot(fl_mv['N'], fl_mv['wall_time'], '--o', label='FLAM_MV')
ax.plot(s_mv['N'], s_mv['wall_time'], '--o', label='STRUMPACK_MV')


ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('N')
ax.set_ylabel('Time (s)')

ax.legend()
plt.savefig('../../images/ch_2/alg_matvec.pdf', format='pdf')
```

```python
fig, ax = plt.subplots()

ax.plot(f['N'], f['max_rss'], '--o', label='FMM3D')
ax.plot(fl['N'], fl['max_rss'], '--o', label='FLAM')
ax.plot(s['N'], s['max_rss'], '--o', label='STRUMPACK')
ax.plot(e['N'], e['max_rss'], '--x', label='ExaFMM')
ax.plot(et['N'], et['max_rss'], '--^', label='ExaFMM-T')



ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('N')
ax.set_ylabel('Peak Memory (GB)')
ax.legend()
plt.savefig('../../images/ch_2/memory.pdf', format='pdf')
```
