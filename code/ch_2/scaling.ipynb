```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()

s = df[df['method']=='strumpack']
s_mv = df[df['method']=='strumpack_mv']
e = df[df['method']=='exafmm']
et = df[df['method']=='exafmmt']
f = df[df['method']=='fmm3d']
fl = df[df['method']=='flam']
fl_mv = df[df['method']=='flam_mv']
```

```python
x = np.array([10**3, 10**4, 10**5, 10**6, 10**7])
y = np.array([10**0, 10**1, 10**2, 10**3, 10**4])
y1 = y*0.3
y2 = y**2*0.3
```

```python
fig, ax = plt.subplots(1,1)


ax.plot(fl['N'], fl['wall_time'], '--8', label='FLAM')
ax.plot(s['N'], s['wall_time'], '--X', label='STRUMPACK')
ax.plot(f['N'], f['wall_time'], '--o', label='FMM3D')
ax.plot(e['N'], e['wall_time'], '--x', label='ExaFMM')
ax.plot(et['N'], et['wall_time'], '--^', label='ExaFMM-T')
ax.plot(x, y1, color='black')
ax.plot(x, y2, color='blue')

ax.text(500012, 1000000, r'$O(N^2)$', fontsize=15, color='blue')

ax.text(500012, 500, r'$O(N)$', fontsize=15)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('N')
ax.set_ylabel('Time (s)')

ax.legend()
plt.savefig('../../images/ch_2/runtime.pdf', dpi=10000, format='pdf')
```

```python
fig, ax = plt.subplots()

ax.plot(fl_mv['N'], fl_mv['wall_time'], '--8', label='FLAM_MV')
ax.plot(s_mv['N'], s_mv['wall_time'], '--X', label='STRUMPACK_MV')

x = np.array([10**3, 10**4, 10**5])
y = np.array([10**-3, 10**-2, 10**-1])
y1 = y
ax.plot(x, y1, color='black')

ax.text(2700, 0.002, r'$O(N)$', fontsize=15)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('N')
ax.set_ylabel('Time (s)')


ax.legend()
plt.savefig('../../images/ch_2/alg_matvec.pdf', dpi=10000, format='pdf')
```
